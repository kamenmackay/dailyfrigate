```python
def concatDirectory(inputdir):
    input_files = sorted(glob.glob(f"{inputdir}/*.mp4"))
    vidstreams = [ ffmpeg.input(x) for x in input_files]
    joined = ffmpeg.concat(*vidstreams, v=1, a=0).node
    filtered_video = (joined[0].filter('mpdecimate', hi=10800, lo=1600, frac=0.1).filter('setpts', 'N/FRAME_RATE/TB'))
    # filtered_video = (joined[0].filter('mpdecimate').filter('setpts', 'N/FRAME_RATE/TB'))
    base_name = inputdir.strip('/').replace('/Volumes/cams/recordings/', '').replace('/', '-')
    output_file = f"{base_name}.mp4"
    output = ffmpeg.output(filtered_video, output_file, vcodec='hevc_videotoolbox', q=16, tag='hvc1',r=15)
    # output = ffmpeg.output(filtered_video, output_file, vcodec='libx265', preset='ultrafast', tag='hvc1')
    ffmpeg.run(output)
```

```python
concatDirectory(testdir)
```

```python
for dt in directories:
    concatDirectory(dt)
```
